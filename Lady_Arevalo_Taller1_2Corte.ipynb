## 1. Ejercicio Fechas Publicación



Teniendo en cuenta que existe una Fecha de Publicación del Proceso y una fecha de última publicación, realice la limpieza de formato de dichas variables. Defina una nueva variable en la que se calcule la diferencia de éstas dos fechas y responda si incluiría las dos fechas dentro de la fuente de datos.

*Utilice estadísticas descriptivas para responder la pregunta*

## **Solución:** 

Procedemos a la limpieza del formato de la variable fecha de Publicación del Proceso y una fecha de última publicación de la base de datos Datos_SECOP:

### 1. Limpieza y ejecución


```python
df_SECOP['Fecha de Publicacion del Proceso'] = pd.to_datetime(df_SECOP['Fecha de Publicacion del Proceso'],format='%m/%d/%Y')
df_SECOP['Fecha de Publicacion del Proceso'].head()
```

```python
df_SECOP['Fecha de Ultima Publicación'] = pd.to_datetime(df_SECOP['Fecha de Ultima Publicación'],format='%m/%d/%Y')
df_SECOP['Fecha de Ultima Publicación'].head()
```

Procedemos a seleccionar las dos columnas con las variables Fecha de Publicacion del Proceso y Fecha de Ultima Publicación y creo una base que se llama datos_Fechas para observar el analisis para las fechas de publicación.

```python
columns = ['Entidad', 'Fecha de Publicacion del Proceso','Fecha de Ultima Publicación']
datos_Fechas = df_SECOP.copy()
datos_Fechas = datos_Fechas[columns]
datos_Fechas
```

```python
datos_Fechas['dif_fechas'] =  (datos_Fechas['Fecha de Publicacion del Proceso'] - datos_Fechas['Fecha de Ultima Publicación']).dt.days 
datos_Fechas
```

Vamos a realizar un analisis descriptivo basico para evaluar nuestra nueva variable y miramos la consistencia 

```python
datos_Fechas['dif_fechas'].value_counts(dropna=False)
```

```python
datos_Fechas['porcentaje'] = (datos_Fechas['dif_fechas'] / 
                  datos_Fechas['dif_fechas'].sum()) * 10
datos_Fechas
```

- Se crea una nueva variable que se llama dif_fechas donde ponemos la diferencia para así evaluar tiempo que tardo desde la Fecha de Publicacion del Proceso hasta la Fecha de Ultima Publicación y observar si tiene alguna diferencia significativa para tomarla o no en nuestro análisis.

- En conclusión, podemos observar que la diferencia entre las dos variables Fecha de Publicacion del Proceso y Fecha de Ultima Publicación no tiene ninguna relevancia la diferencia es de 1 día para 125895 entidades y tenemos cero diferencias en 750871 entidades, la diferencia porcentual es del 0.000079 % para todos los casos con 1 día de diferencia.

- Si incluiría las dos variables ya que hay varios datos que tiene n1 día de diferencia y veo que para el lector es importante saber que no hay una diferencia significativa en las fechas de Fecha de Publicación del Proceso y Fecha de Ultima Publicación.
